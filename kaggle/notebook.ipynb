```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../dataset/ghostindex_benchmarks.csv')
print('Dataset shape:', df.shape)
print(df.head(10))
```

```python
violation_counts = df['violation_type'].value_counts()
colors = ['#ef4444','#f97316','#8b5cf6','#3b82f6','#10b981']
plt.figure(figsize=(8,5))
plt.pie(violation_counts, labels=violation_counts.index, colors=colors, autopct='%1.0f%%')
plt.title('BHMarketer GhostIndex - Violation Type Distribution', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
signal_cols = ['expired_content_score','misinformation_score','policy_violation_score','violent_content_score','compliance_risk_score']
signal_avgs = df[signal_cols].mean()
plt.figure(figsize=(12,5))
plt.bar(signal_avgs.index, signal_avgs.values, color=colors)
plt.title('BHMarketer GhostIndex - Average Detection Signal Scores', fontsize=14, fontweight='bold')
plt.xlabel('Detection Signal')
plt.ylabel('Average Score (0-100)')
plt.xticks(rotation=20)
plt.axhline(y=60, color='red', linestyle='--', alpha=0.5, label='High Risk threshold')
plt.legend()
plt.tight_layout()
plt.show()
```
